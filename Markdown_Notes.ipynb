### üìå **Jupyter Notebook Markdown Cheatsheet - Hinglish Notes**  

Jupyter Notebook me **Markdown** ka use text formatting, headings, lists, links, images aur code blocks likhne ke liye hota hai. Yeh data science aur machine learning me **notes likhne aur reports banane** ke liye kaafi useful hai.  

---

## **1Ô∏è‚É£ Headings (Title aur Subheadings)**
Markdown me **heading likhne ke liye `#` ka use** hota hai.  
Jitne zyada `#`, utni chhoti heading size.  

```markdown
# Sabse Badi Heading (H1)
## Doosri Badi Heading (H2)
### Teesri Badi Heading (H3)
#### Chauthi Heading (H4)
##### Paanchvi Heading (H5)
###### Chhathi Heading (H6)
```

‚û°Ô∏è Example:  
# Ye H1 Heading Hai  
## Ye H2 Heading Hai  
### Ye H3 Heading Hai  

---

## **2Ô∏è‚É£ Bold aur Italic Text**  
Kisi text ko **bold** ya *italic* me likhne ke liye `**` ya `*` ka use hota hai.  

```markdown
**Yeh Bold Text Hai**  
*Yeh Italic Text Hai*  
***Yeh Bold + Italic Hai***  
```

‚û°Ô∏è Example:  
**Yeh Bold Hai**  
*Yeh Italic Hai*  
***Yeh Bold + Italic Hai***  

---

## **3Ô∏è‚É£ Lists (Ordered & Unordered)**  
### üìç **Unordered List (Bullet Points)**
Bullet points banane ke liye `-`, `+`, ya `*` ka use kar sakte hain.  

```markdown
- Apple  
- Mango  
- Banana  

+ Car  
+ Bike  
+ Bus  

* Python  
* Java  
* C++  
```

‚û°Ô∏è Example:  
- Apple  
- Mango  
- Banana  

+ Car  
+ Bike  
+ Bus  

* Python  
* Java  
* C++  

### üìç **Ordered List (Numbering)**
```markdown
1. First Item  
2. Second Item  
3. Third Item  
```

‚û°Ô∏è Example:  
1. First Item  
2. Second Item  
3. Third Item
